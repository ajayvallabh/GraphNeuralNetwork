# Multiple Edge Attribute

Create adjacency matrices for multiple edge attributes with structured dtypes:

```python
import numpy as np

G = nx.Graph()
G.add_edge(0, 1, weight=10)
G.add_edge(1, 2, cost=5)
G.add_edge(2, 3, weight=3, cost =-4.0)
dtype = np.dtype([("weight", int), ("cost", float)])
A = nx.to_numpy_array(G, dtype=dtype, weight=None)
print("weight ------")
print(A["weight"])
print("cost ------")
print(A["cost"])
nx.draw(G)
print(G.edges)
print(G.nodes)
```

running adjacency representation of graph as a dictionary of dictionaries.

```python
G = nx.Graph()
G.add_edges_from(
    [
        (0, 1, {'weight': 1.0}),
        (1, 2, {'weight': 2.0}),
        (2, 0, {'weight': 1.0}),
    ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("----------------------")
d = nx.to_dict_of_dicts(G)
print(d)
print(d[1][2]['weight'])
```

# MultiGraph

```python
G = nx.MultiGraph()
G.add_edge(0, 1, key ='a', weight = 1.0)
G.add_edge(0, 1, key ='b', weight = 5.0)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("---------------")
d = nx.to_dict_of_dicts(G)
print(d)
print(d[0][1]['b']['weight'])
```

```python
G = nx.Graph(
    [
    ("A","B", {"cost": 1, "weight":7}),
    ("C","E", {"cost": 9, "weight":10}),
    ]
  )

nx.draw(G)
print(G.edges)
print(G.nodes)
print("------------------")
df = nx.to_pandas_edgelist(G)
print(df)
print("------------------")
df = nx.to_pandas_edgelist(G, nodelist=["B", "E"])
print(df)
print("------------------")
df = nx.to_pandas_edgelist(G, nodelist=["A", "C"])
print(df)
print("------------------")
df[["source", "target", "cost", "weight"]]
```

printing Adjacency Matrix

```python
G = nx.Graph(
    [
        ("A","B", {"cost": 1, "weight":7}),
        ("c", "E", {"cost":9, "weight":10}),
    ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("-------------------")
df = nx.to_pandas_edgelist(G)
print(df)
df = nx.to_pandas_edgelist(G, nodelist=["B", "E"])
print(df)
print("------------------")
df = nx.to_pandas_edgelist(G, nodelist=["A", "C"])
print(df)
print("------------------")
df[["source", "target", "cost", "weight"]]
print("------------------")
A = nx.adjacency_matrix(G)
print(A.todense())
```

```python
G = nx.MultiGraph([('A','B', {'cost':1}),('A','B', {'cost':9})])
nx.draw(G)
print(G.edges)
print(G.nodes)
print("-----------------")
df = nx.to_pandas_edgelist(G, nodelist=['A','C'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
print("-----------------")
df = nx.to_pandas_edgelist(G, nodelist=['B'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
```

# Sudoku Graph

The sudoku graph is a graph with n^4 vertices, corresponding to the cells of an n^2 by n^2 grid. Two distinct vertices are adjacent if and only if they belong to the same row, column, or n-by-b box.

Parameters: n integer
The order of the Sudoku graph, equal to the square root of the number of rows. The default is three

```python
# Sudoku graph with default 3x3 = 9 nodes
G = nx.sudoku_graph()
print(G.number_of_nodes())
print(G.number_of_edges())
print("-----------------")
A = nx.adjacency_matrix(G)
print(A.todense())
nx.draw(G)
```

```python
print(G.is_directed())
print(G.is_multigraph())
```

```python
# Creating a 2x2 sudoku graph
G = nx.sudoku_graph(2)
print(G.number_of_nodes())
print(G.number_of_edges())
print("-----------------")
A = nx.adjacency_matrix(G)
print(A.todense())
nx.draw(G)
```
